# Exploratory Data Analysis (EDA)

This notebook explores the LiDAR point cloud dataset for RiDAR_net project.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------------
# 1. LOAD DATA (CSV version)
# -----------------------------------
csv_file = "data/raw/Lidar_dataset_original_4scales.csv"
df = pd.read_csv(csv_file, low_memory=False)

print("Dataset shape:", df.shape)
print(df.head())

# -----------------------------------
# 2. EXTRACT POINT FEATURES
# -----------------------------------
x_cols = [c for c in df.columns if c.startswith("x")]
y_cols = [c for c in df.columns if c.startswith("y")]
z_cols = [c for c in df.columns if c.startswith("z")]
i_cols = [c for c in df.columns if c.startswith("i")]

# Take first cluster for visualization
sample = df.iloc[0]

x = sample[x_cols].values
y = sample[y_cols].values
z = sample[z_cols].values
i = sample[i_cols].values

# Remove zero padding
mask = (x != 0) | (y != 0) | (z != 0)
x, y, z, i = x[mask], y[mask], z[mask], i[mask]

# -----------------------------------
# 3. BASIC STATS
# -----------------------------------
print("\n=== Basic Stats ===")
print("Points:", len(x))
print("Z min/max:", z.min(), z.max())
print("Intensity min/max:", i.min(), i.max())

# -----------------------------------
# 4. 3D SCATTER PLOT
# -----------------------------------
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

sc = ax.scatter(x, y, z, c=z, s=2)
plt.colorbar(sc, label="Elevation (Z)")
ax.set_title("3D Point Cloud (Sample Cluster)")
plt.show()

# -----------------------------------
# 5. TOP VIEW (XY)
# -----------------------------------
plt.figure(figsize=(6,6))
plt.scatter(x, y, c=z, s=2)
plt.colorbar(label="Elevation")
plt.title("Top View (XY)")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

# -----------------------------------
# 6. HEIGHT DISTRIBUTION
# -----------------------------------
plt.figure()
plt.hist(z, bins=50)
plt.title("Z Distribution (Height)")
plt.xlabel("Z")
plt.ylabel("Frequency")
plt.show()

# -----------------------------------
# 7. INTENSITY DISTRIBUTION
# -----------------------------------
plt.figure()
plt.hist(i, bins=50)
plt.title("Intensity Distribution")
plt.xlabel("Intensity")
plt.ylabel("Frequency")
plt.show()

# -----------------------------------
# 8. CLASS DISTRIBUTION
# -----------------------------------
if "label" in df.columns:
    print("\n=== Class Distribution ===")
    print(df["label"].value_counts())

    df["label"].value_counts().plot(kind="bar")
    plt.title("Class Distribution")
    plt.show()
```
